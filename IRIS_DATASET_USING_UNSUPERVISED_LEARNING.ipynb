```python
features = df.loc[:,["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=99) 
```

```python
## WCSS MEANS Within-Cluster Sum of Square. SUM OF SQUARE DISTANCE BETWEEN POINTS IN A CLUSTER AND THE CLUSTER CENTROID##
```

```python
from sklearn.cluster import KMeans
wcss = []

for k in range(1,5):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(features)
    wcss.append(kmeans.inertia_)


plt.figure(figsize=(5,5))
plt.title("WCSS / K Chart", fontsize=18)
plt.plot(range(1,5),wcss,"-o")
plt.grid(True)
plt.xlabel("Amount of Clusters",fontsize=14)
plt.ylabel("Inertia",fontsize=14)
plt.xticks(range(1,20))
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc_cluster = AgglomerativeClustering(n_clusters=99) 
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

merg = linkage(features,method="ward")

plt.figure(figsize=(1,5))
dendrogram(merg, leaf_rotation=90)
plt.xlabel("data points")
plt.ylabel("euclidian distance")

plt.suptitle("DENDROGRAM",fontsize=18)
plt.show()
```
